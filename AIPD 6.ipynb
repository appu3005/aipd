```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies = pd.read_csv("/home/admin1/Downloads/imdb_top_1000.csv")
movies = movies[['Series_Title','Genre','IMDB_Rating']]
movies.head()
```

```python
tfidf = TfidfVectorizer()

genre_matrix = tfidf.fit_transform(movies['Genre'])
```

```python
model = NearestNeighbors(
        metric='cosine',
        algorithm='brute',
        n_neighbors=6)

model.fit(genre_matrix)
```

```python
def recommend_movie(movie_name):

    index = movies[movies['Series_Title'] == movie_name].index[0]

    distances, indices = model.kneighbors(
        genre_matrix[index],
        n_neighbors=6)

    print("Recommended Movies:\n")

    for i in indices.flatten()[1:]:
        print(movies.iloc[i]['Series_Title'])
```

```python
recommend_movie("The Dark Knight")
```

### Naive Bayes Model for the Movie Recommendation

```python
movies = pd.read_csv("/home/admin1/Downloads/imdb_top_1000.csv")

movies = movies[['Series_Title','Overview','IMDB_Rating']]

movies.dropna(inplace=True)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# create label column
movies['label'] = movies['IMDB_Rating'].apply(lambda x: 1 if x >= 8 else 0)

# convert text to numbers
tfidf = TfidfVectorizer(stop_words='english')

X = tfidf.fit_transform(movies['Overview'])

y = movies['label']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test,y_pred))
print("Precision:", precision_score(y_test,y_pred))
print("Recall:", recall_score(y_test,y_pred))
```

```python
def recommend_movie(description):

    text = tfidf.transform([description])

    prediction = model.predict(text)

    if prediction[0] == 1:
        print("This movie is likely to be Recommended")
    else:
        print("This movie may not be Recommended")
```
